```python
#Exploring data from different angles

comp_types = df.groupby(['fado_type', 'allegation'])['complaint_id'].count().reset_index()
# comp_types
comp_types_s = df.groupby(['fado_type']).complaint_id.count().reset_index()
comp_types_s
# print(type(comp_types))
```

```python
# Quick barplot reflecting groupby code
# sns.set(style="whitegrid", color_codes=True)
sns.set_style('ticks')
ax = sns.barplot(x = 'fado_type', y = 'complaint_id', data = comp_types_s, palette = ["red", "green", "blue", "yellow"], ci = False)
# "Error bars" signify confidence interval can be removed with ci = False
```

```python
# Switching axes to check for readability 
# sns.set(style="whitegrid", color_codes=True)
sns.set(rc={'figure.figsize': (16,3)})
ax = sns.barplot(x = 'complaint_id', y = 'fado_type', data = comp_types_s, palette = ["red", "green", "blue", "yellow"], orient = 'h', ci = False)
```

```python
# Abuse of authority as a broader category has the most cumulative complaints, but let's take a closer look at complaints over time 
sns.set_style('ticks')
ax = sns.lineplot(x = 'year_received', y = 'complaint_id', data = df, style = 'fado_type', hue = 'fado_type', 
                  palette = ["red", "green", "blue", "yellow"], ci = False, markers = True)
# TODO: Redo graph using more specific parameters
```

```python
# Creating police_id column, which combines unique_mos_id, first_name, and last_name for easier grouping

df['police_id'] = df[df.columns[0:3]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1)
df
```

```python
# Grouping by police_id

unique_grouping = df.groupby(['police_id']).complaint_id.count().reset_index()
unique_grouping
```

```python
# Taking the average of the count of complaint_id to see what the average amount of complaints per officer is.

unique_grouping['complaint_id'].mean()
```

```python
# Selecting the cops with complaints greater than the average

unique_grouping.loc[unique_grouping['complaint_id'] > 8]
```
