```python
#Exploring data from different angles

comp_types = df.groupby(['fado_type', 'allegation'])['complaint_id'].count().reset_index()
# comp_types
comp_types_s = df.groupby(['fado_type']).complaint_id.count().reset_index()
comp_types_s
# print(type(comp_types))
```

```python
# Quick barplot reflecting groupby code
# sns.set(style="whitegrid", color_codes=True)
sns.set_style('ticks')
ax = sns.barplot(x = 'fado_type', y = 'complaint_id', data = comp_types_s, palette = ["red", "green", "blue", "yellow"], ci = False)
# "Error bars" signify confidence interval can be removed with ci = False
```

```python
# Switching axes to check for readability 
# sns.set(style="whitegrid", color_codes=True)
sns.set(rc={'figure.figsize': (16,3)})
ax = sns.barplot(x = 'complaint_id', y = 'fado_type', data = comp_types_s, palette = ["red", "green", "blue", "yellow"], orient = 'h', ci = False)
```

```python
# Creating police_id column, which combines unique_mos_id, first_name, and last_name for easier grouping

df['police_id'] = df[df.columns[0:3]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1)
df
```

```python
# Grouping by police_id

unique_grouping = df.groupby(['police_id']).complaint_id.count().reset_index()
unique_grouping
```

```python
# Taking the average of the count of complaint_id to see what the average amount of complaints per officer is.

unique_grouping['complaint_id'].mean()
```

```python
# Selecting the cops with complaints greater than the average

prob_cops = unique_grouping.loc[unique_grouping['complaint_id'] > 8]
# print(type(prob_cops))
prob_cops
```

```python
# Saving prob_cops to csv to share with team

prob_cops.to_csv('problem_cops.csv')
```

```python
# Donut chart of outcomes: substantiated, unsubstantiated, exonerated

mm_donut = df.copy()
# mm_donut

# Replacing any Substantiated subcat w/ 'Substantiated' for cleaner graph

mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Charges)'],'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Command Discipline A)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Command Discipline B)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Command Discipline)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Command Lvl Instructions)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Formalized Training)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (Instructions)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (MOS Unidentified)'], 'Substantiated')
mm_donut['board_disposition'] = mm_donut['board_disposition'].replace(['Substantiated (No Recommendations)'], 'Substantiated')
mm_donut
```

```python
# Group by board_disposition and count complaint_id

mm_donut.groupby(['board_disposition'])['complaint_id'].count().reset_index()
```

```python
# Donut chart (continued)

# create data
names=[r'Exonerated (29 %)', r'Substantiated (25 %)', 
r'Unsubstantiated (46 %)']
size=[9609,8301,15448]
 
# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.7, color='white')

# Give color names
# plt.pie(size, labels=names, colors=['red','green','blue'])
# p=plt.gcf()
# p.gca().add_artist(my_circle)
# plt.show()

# Label distance: gives the space between labels and the center of the pie
# plt.pie(size, labels=names, labeldistance=0.45)
# p=plt.gcf()
# p.gca().add_artist(my_circle)
# plt.show()
 
# Label color
# plt.rcParams['text.color'] = 'red'
# plt.pie(size, labels=names)
# p=plt.gcf()
# p.gca().add_artist(my_circle)
# plt.show()

# Custom wedges
# plt.pie(size, labels=names, wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
# p=plt.gcf()
# p.gca().add_artist(my_circle)
# plt.show()

plt.rcParams['text.color'] = 'black'
plt.pie(size, labels=names, colors=['red','green','blue'], labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('CCRB Findings')

plt.savefig('donut.png', dpi=100)

plt.show()
```
